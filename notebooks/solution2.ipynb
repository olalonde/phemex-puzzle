```python
strings = ['ETH', 'BTC', 'XRP', 'Phemex', '957496696762772407663', 'satoshi', 'dorian', 'nakamoto']
```

```python
prime21 = 957496696762772407663
```

```python
def sha256(s):
    h = s.encode()
    h = hashlib.sha256(h).digest()
    return int(h.hex(), 16)
def double_sha256(s):
    h = s.encode()
    h = hashlib.sha256(h).digest()
    h = hashlib.sha256(h).digest()
    return int(h.hex(), 16)
def ripemd(s):
    h = s.encode()
    h2 = hashlib.new('ripemd160')
    h2.update(h)
    h = h2.digest()
    return int(h.hex(), 16)
def double_sha256_ripemd(s):
    h = s.encode()
    h = hashlib.sha256(h).digest()
    h = hashlib.sha256(h).digest()
    h2 = hashlib.new('ripemd160')
    h2.update(h)
    h = h2.digest()
    return int(h.hex(), 16)
def sha256_ripemd(s):
    h = s.encode()
    h = hashlib.sha256(h).digest()
    h2 = hashlib.new('ripemd160')
    h2.update(h)
    h = h2.digest()
    return int(h.hex(), 16)
def ripemd_sha256(s):
    h = s.encode()
    h2 = hashlib.new('ripemd160')
    h2.update(h)
    h = h2.digest()
    h = hashlib.sha256(h).digest()
    return int(h.hex(), 16)
def ripemd_double_sha256(s):
    h = s.encode()
    h = hashlib.sha256(h).digest()
    h = hashlib.sha256(h).digest()

    h2 = hashlib.new('ripemd160')
    h2.update(h)
    h = h2.digest()
    return int(h.hex(), 16)
```
